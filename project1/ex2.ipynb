```python
#basic imports
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#load the datasets
fileTrainImages = open("abalone.data", "r")
DataList=[]
LabelsList=[]
for line in fileTrainImages:
    items=line.split(",")
    if(items[0]=='M'):
        items[0]='0'
    elif(items[0]=='F'):
        items[0]='1'
    elif(items[0]=='I'):
        items[0]='2'
    DataList.append(items[:-1])
    LabelsList.append(items[-1].split("\n")[0])
Data=np.array(DataList).astype(float)
Labels=np.array(LabelsList).astype(float)
print Data
print Data.shape
print Labels
print Labels.shape
```

```python
#linear regression
from sklearn import linear_model

kfLinear = KFold(n_splits=10,shuffle=True)
kfLinear.get_n_splits(Data)

iteration=1
varianceScores=[]
meanSquaredErrors=[]
for train_index, test_index in kfLinear.split(Data):   
    DataTrain, DataTest = Data[train_index], Data[test_index]
    LabelsTrain, LabelsTest = Labels[train_index], Labels[test_index]
    lm = linear_model.LinearRegression(fit_intercept=True,normalize=True)
    lm.fit(DataTrain, LabelsTrain)
    predictedLabels = lm.predict(DataTest)
    print "Variance score on "+str(iteration)+" iteration = " +str(r2_score(LabelsTest, predictedLabels))
    print "Mean squared error on "+str(iteration)+" iteration = " +str(mean_squared_error(LabelsTest, predictedLabels))
    varianceScores.append(r2_score(LabelsTest, predictedLabels))
    meanSquaredErrors.append(mean_squared_error(LabelsTest, predictedLabels))
    iteration=iteration+1
print "Variance's Mean score = "+str(sum(varianceScores)/len(varianceScores))
print "Mean of Mean's Squared error  = "+str(sum(meanSquaredErrors)/len(meanSquaredErrors))
```

```python
#polynomial regression
from sklearn.preprocessing import PolynomialFeatures

kfPolynomial = KFold(n_splits=10,shuffle=True)
kfPolynomial.get_n_splits(Data)
 
iteration=1
varianceScores=[]
meanSquaredErrors=[]
for train_index, test_index in kfPolynomial.split(Data):   
    DataTrain, DataTest = Data[train_index], Data[test_index]
    LabelsTrain, LabelsTest = Labels[train_index], Labels[test_index]
    polyReg = PolynomialFeatures(degree=2)
    DataPolyTrain = polyReg.fit_transform(DataTrain)
    DataPolyTest = polyReg.fit_transform(DataTest)
    pm = linear_model.LinearRegression()
    pm.fit(DataPolyTrain, LabelsTrain)
    predictedLabels = pm.predict(DataPolyTest)
    print "Variance score on "+str(iteration)+" iteration = " +str(r2_score(LabelsTest, predictedLabels))
    print "Mean squared error on "+str(iteration)+" iteration = " +str(mean_squared_error(LabelsTest, predictedLabels))
    varianceScores.append(r2_score(LabelsTest, predictedLabels))
    meanSquaredErrors.append(mean_squared_error(LabelsTest, predictedLabels))
    iteration=iteration+1
print "Mean Variance's score = "+str(sum(varianceScores)/len(varianceScores))
print "Mean of Mean's Squared error  = "+str(sum(meanSquaredErrors)/len(meanSquaredErrors))
```

```python
#lasso
from sklearn import linear_model

kfLasso = KFold(n_splits=10,shuffle=True)
kfLasso.get_n_splits(Data)

iteration=1
varianceScores=[]
meanSquaredErrors=[]
for train_index, test_index in kfLasso.split(Data):
    DataTrain, DataTest = Data[train_index], Data[test_index]
    LabelsTrain, LabelsTest = Labels[train_index], Labels[test_index] 
    las = linear_model.Lasso(alpha=0.001)
    las.fit(DataTrain, LabelsTrain)
    predictedLabels = las.predict(DataTest)
    print "Variance score on "+str(iteration)+" iteration = " +str(r2_score(LabelsTest, predictedLabels))
    print "Mean squared error on "+str(iteration)+" iteration = " +str(mean_squared_error(LabelsTest, predictedLabels))
    varianceScores.append(r2_score(LabelsTest, predictedLabels))
    meanSquaredErrors.append(mean_squared_error(LabelsTest, predictedLabels))
    iteration=iteration+1
print "Mean Variance's score = "+str(sum(varianceScores)/len(varianceScores))
print "Mean of Mean's Squared error  = "+str(sum(meanSquaredErrors)/len(meanSquaredErrors))
```

```python
#mlp regressor
from sklearn.neural_network import MLPRegressor

kfMlp = KFold(n_splits=10,shuffle=True)
kfMlp.get_n_splits(Data)

iteration=1
varianceScores=[]
meanSquaredErrors=[]
for train_index, test_index in kfMlp.split(Data):  
    DataTrain, DataTest = Data[train_index], Data[test_index]
    LabelsTrain, LabelsTest = Labels[train_index], Labels[test_index]
    mlp = MLPRegressor(hidden_layer_sizes=(5, ), activation="tanh", solver="lbfgs")
    mlp.fit(DataTrain, LabelsTrain)
    predictedLabels = mlp.predict(DataTest)
    print "Variance score on "+str(iteration)+" iteration = " +str(r2_score(LabelsTest, predictedLabels))
    print "Mean squared error on "+str(iteration)+" iteration = " +str(mean_squared_error(LabelsTest, predictedLabels))
    varianceScores.append(r2_score(LabelsTest, predictedLabels))
    meanSquaredErrors.append(mean_squared_error(LabelsTest, predictedLabels))
    iteration=iteration+1
print "Mean Variance's score = "+str(sum(varianceScores)/len(varianceScores))  
print "Mean of Mean's Squared error  = "+str(sum(meanSquaredErrors)/len(meanSquaredErrors))
```

```python
#gaussian process 
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF,DotProduct

kfGp = KFold(n_splits=10,shuffle=True)
kfGp.get_n_splits(Data)

iteration=1
varianceScores=[]
meanSquaredErrors=[]
kernels = [DotProduct(),RBF(length_scale=0.01)]
for i, kernel in enumerate(kernels):
    for train_index, test_index in kfGp.split(Data):
        if(iteration==11):
            iteration=1
            varianceScores=[]
            meanSquaredErrors=[] 
            print "       ----------------------------------           "
        DataTrain, DataTest = Data[train_index], Data[test_index]
        LabelsTrain, LabelsTest = Labels[train_index], Labels[test_index]
        gpr = GaussianProcessRegressor(kernel=kernel,random_state=0)
        gpr.fit(DataTrain, LabelsTrain)
        predictedLabels = gpr.predict(DataTest)
        print "Variance score on "+str(iteration)+" iteration = " +str(r2_score(LabelsTest, predictedLabels))
        print "Mean squared error on "+str(iteration)+" iteration = " +str(mean_squared_error(LabelsTest, predictedLabels))
        varianceScores.append(r2_score(LabelsTest, predictedLabels))
        meanSquaredErrors.append(mean_squared_error(LabelsTest, predictedLabels))
        iteration=iteration+1        
    print "Mean Variance's score = "+str(sum(varianceScores)/len(varianceScores))   
    print "Mean of Mean's Squared error  = "+str(sum(meanSquaredErrors)/len(meanSquaredErrors))
```
